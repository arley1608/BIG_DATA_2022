```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Scatter plot (diagramas de dispersión)
x = np.arange(1, 20)
y = x**2
y1 = x**2 + 5*x

plt.scatter(x, y, s = 10, color = 'red', marker = '*', label = 'x^2')
plt.scatter(x, y1, s = 10, color = 'blue', marker = 'o', label = 'x^2 + 5x')
plt.legend(title = 'Funciones')
plt.title('Tipos de Funciones', fontsize = 20)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
# plt.xlim(0, 5)

plt.text(5, 300, 'este texto', color = 'brown')

# agregar coordenadas de los marcadores en texto
for val_x in x:
    plt.text(val_x, val_x**2, f'({val_x}, {val_x**2})', fontsize = 5)
```

```python
# Histograma

media = 20
std = 3
numeros = np.random.normal(media, std, 10000)

plt.hist(numeros, bins = 20, cumulative = False, density = False);

plt.title('Distribución Normal', fontsize = 20)
plt.xlabel('Números')
plt.ylabel('Frecuencia')
plt.grid()
```

```python
plt.boxplot(numeros, vert = False);
```
